## **Pasos**

Sea $X$ el problema a resolver.

Las soluciones están representadas como tira de bits.

$\quad$ **1.** Comenzar con una población $P$ generada aleatoriamente de $n$ cromosomas de $k$ bits.

$\quad$ **2.** Calcular la aptitud $f(x)$ para cada cromosoma $x$ de $P$. 

$\quad$ **3.** Repetir los siguientes pasos hasta que se hayan creado $n$ descendientes:

$\quad\quad$ **a.** Seleccionar un par de cromosomas de $P$, siendo la _probabilidad de selección_ una función creciente de la $aptitud$.

$\quad\quad\quad$ **a$_1$.** Con **probabilidad pc** (probabilidad de cruce), cruzar el par en un punto elegido aleatoriamente para formar dos descendientes (si no tiene lugar ningún cruce, formar dos descendientes que sean copias exactas de sus respectivos padres) y agregarlos a una nueva población $P’$.

$\quad\quad\quad$ **a$_2$.** Mutar los dos descendientes en cada lugar con **probabilidad pm** (probabilidad de mutación) y colocar los cromosomas resultantes en la nueva población $P’$.

$\quad\quad$ **b.**  Si $n$ es impar, se puede rechazar aleatoriamente a un miembro de la nueva población. 

$\quad$ **4.** Reemplazar la población actual $P$ con la nueva $P’$. 

$\quad$ **5.** Volver al paso $2.$.

## **Código: El problema de las $8$ reinas**

Se quiere poder colocar el mayor número posible de reinas en una tablero de ajedrez sin que ninguna reina pueda comerse a otra.

image.png

Como las reinas pueden avanzar a lo largo de toda una fila, por lógica el mayor número posible serán a lo sumo $8$ (hay $8$ filas en el tablero).

Se instala $pyeasyga$ y se importan las bibliotecas necesarias.

```python
!pip install pyeasyga
```

```python
import random
from pyeasyga import pyeasyga
```

Se define cromosoma inicial.

```python
# setup seed data
seed_data = [0, 1, 2, 3, 4, 5, 6, 7]
```

Se inicializa un algoritmo genético.

```python
# initialise the GA
ga = pyeasyga.GeneticAlgorithm(seed_data,
                               population_size=200,
                               generations=100,
                               crossover_probability=0.8,
                               mutation_probability=0.2,
                               elitism=True,
                               maximise_fitness=False)
```

Se define una función para crear un individuo.

```python
# define and set function to create a candidate solution representation
def create_individual(data):
    individual = data[:] #make a copy
    random.shuffle(individual) #change the order
    return individual

ga.create_individual = create_individual
```

Se define la **cruza**.

```python
# define and set the GA's crossover operation
def crossover(parent_1, parent_2):
    crossover_index = random.randrange(1, len(parent_1)) #The randrange() method returns a randomly selected element from the specified range.    =>  random.randrange(start, stop, step)
    child_1a = parent_1[:crossover_index]  #return array from start to index, from "parent 1_"
    child_1b = [i for i in parent_2 if i not in child_1a] # second part of the array taken from parent_2
    child_1 = child_1a + child_1b 

    child_2a = parent_2[crossover_index:]
    child_2b = [i for i in parent_1 if i not in child_2a]
    child_2 = child_2a + child_2b

    return child_1, child_2

ga.crossover_function = crossover
```

Se define la **mutación**.

```python
# define and set the GA's mutation operation
def mutate(individual):
    mutate_index1 = random.randrange(len(individual))
    mutate_index2 = random.randrange(len(individual))
    index_val1 = individual[mutate_index1]
    index_val2 = individual[mutate_index2]
    index_val1, index_val2 = index_val2, index_val1 # switch two array elements

ga.mutate_function = mutate
```

Se define la **selección**.

```python
# define and set the GA's selection operation
def selection(population):
    return random.choice(population) #returns a random element

ga.selection_function = selection 
```

Se define la **función aptitud**.

```python
# define a fitness function
def fitness(individual, data):
    collisions = 0
    for item in individual:
        item_index = individual.index(item)
        for elem in individual:
            elem_index = individual.index(elem)
            if item_index != elem_index:
                if item - (elem_index - item_index) == elem \
                        or (elem_index - item_index) + item == elem:
                    collisions += 1
    return collisions

ga.fitness_function = fitness       # set the GA's fitness function
```

Se ejecuta:

```python
ga.run()                            # run the GA
```
